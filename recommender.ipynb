## Make custom knn class

```python
from sklearn.neighbors import NearestNeighbors
import os
```

```python
class MyKNN:
    def __init__(self, path_to_data):
        self.data_path = path_to_data
        self.model = NearestNeighbors()
    
    def set_model_params(self, n_neighbors=5, algorithm="auto", metric="euclidean", n_jobs=None):
        '''
            Parameters
            ----------
            n_neighbors: int, optional (default = 5)
            
            algorithm: {'auto', 'ball_tree', 'kd_tree', 'brute'}, optional
            
            metric: string or callable, default 'minkowski', or one of
                ['cityblock', 'cosine', 'euclidean', 'l1', 'l2', 'manhattan']
            
            n_jobs: int or None, optional (default=None)
        '''
        if n_jobs and (n_jobs > 1 or n_jobs == -1):
            os.environ['JOBLIB_TEMP_FOLDER'] = '/tmp'
        self.model.set_params(**{
            "n_neighbors": n_neighbors,
            "algorithm": algorithm,
            "metric": metric,
            "n_jobs": n_jobs
        })
    
    def _inference(self, model, data, mobile_index, n_recommendations):
        model.fit(data)
        distances, indices = model.kneighbors([data.iloc[mobile_index, :]], n_neighbors= n_recommendations+1)
        raw_recommends = sorted(
            list(
                zip(
                    indices.squeeze().tolist(),
                    distances.squeeze().tolist()
                )
            ), key=lambda x: x[1]
        )[:0:-1]
        return raw_recommends
    
    def get_data(self, index):
        data = read_excel(self.data_path, sheet_name=0, na_values='n/a')
        for column in ["id", "date_rev", "date_buy", "pros", "shown", "cons", "brand_pros", "brand_cons", "id.1", "uname", "email", "uavatar", "youtube", "approved", "twitter", "facebook"]:
            data.drop(column, axis=1, inplace=True)
        data["mobile"] = data["brand"] + " " + data["product"]
        data.drop("product", axis=1, inplace=True)
        data.drop("brand", axis=1, inplace=True)
        data.drop(index, axis=0, inplace=True)
        return data
    
    def make_recommendations(self, index, n_recommendations):
        data = self.get_data(index)
        model_data = data.iloc[:, 1:-1]
        raw_recommends = sorted(self._inference(self.model, model_data, index, n_recommendations), key= lambda x: x[1])
        print(f"Recommendation for '{data.iloc[index, 9]}' ({index})")
        for rec in raw_recommends:
            print(f"{data.iloc[rec[0], 9]} ({rec[0]} with distance {rec[1]})")
        
```

```python
recommender = MyKNN("test.xlsx")
recommender.set_model_params(2, 'auto', 'cosine', -1)
recommender.make_recommendations(0, 4)
```
