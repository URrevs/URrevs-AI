```python
from secrets import choice
import pandas as pd
```

```python
possibilites = [0, 0, 0.1, 0.25, 0, 0.1 , 0.35, 0.5, 0, 0, 0.5, 0, 0.75, 0, 0, 0, 0.85, 0.3, 0, 0, 0, 0.9, 1, 0]
user_id=[]
for i in range(1, 101):
    for j in range(1, 101):
       user_id.append(f'u{i}') 
  
item_id = []
for i in range(1, 101):
    for j in range(1, 101):
       item_id.append(f'i{j}') 

rating=[choice(possibilites) for i in range(1, 10001)]


data = {'user_id':user_id,
        'item_id':item_id,
         'rating':  rating}
User2Review = pd.DataFrame(data)
User2Review
User2Review.to_csv('User2reviwe.csv')
```
