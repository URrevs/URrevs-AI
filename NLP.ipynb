```python
from nltk.corpus import stopwords

more_stopwords = ['آمين', 'أب', 'أخ', 'أفعل', 'أفعله', 'ؤلاء', 'إل', 'إم', 'ات', 'اتان', 'ارتد', 'ان', 'انفك', 'برح', 'تان', 'تبد', 'تحو', 'تعل', 'حد', 'حم', 'حي', 'خب', 'ذار', 'سيما', 'صه', 'ظل', 'ظن', 'عد', 'قط', 'مر', 'مكان', 'مكانكن', 'نب', 'هات', 'هب', 'واها', 'وراء']
stopwords_all = stopwords.words("Arabic") + stopwords.words("English") + more_stopwords

for i in tokenized:
    if i not in stopwords_all:
        print(i)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np

data = pd.read_excel("../Data/data mobiles reviews.xlsx", sheet_name=0, na_values='n/a')
all_sentences = np.array(data["pros"])

vectorizer = TfidfVectorizer(stop_words= stopwords_all)
tfidf = vectorizer.fit_transform(all_sentences)
vectorizer.get_feature_names_out().shape
```

```python
tfidf.shape
```

```python
pd.DataFrame(tfidf.toarray())

arr = np.average(tfidf.toarray(), axis=1)

lst = {}
for i, j in enumerate(arr):
    lst[i] = j

# print(lst)

lst = sorted(lst.items(), key= lambda x:x[1], reverse = True)
print(lst)

len(lst)
```
